```python
df = pd.read_csv('GCB2022v27_MtCO2_flat.csv')
df.head(10)
```

```python
df.count()
```

```python
df.info()
```

```python
year2000_df = df.loc[df['Year'] >= 2000]
year2000_df.head()
```

```python
year2000_df.isna().any()
```

```python
#ISO codes we can fix for the countries
year2000_df[year2000_df['ISO 3166-1 alpha-3'].isna() == True]
```

```python
countries = list(year2000_df[year2000_df['ISO 3166-1 alpha-3'].isna() ==True]['Country'].unique())

len(countries)
```

```python
#drop the 6 countries
cleaned_data_df = year2000_df[~year2000_df['Country'].isin(countries)]
```

```python
cleaned_data_df.fillna(0, inplace=True)

cleaned_data_df.head(10)
```

```python
#check for NaN values
cleaned_data_df.isna().any()
```

```python
cleaned_data_df.columns
```

```python
cleaned_data_df.rename(columns= {'ISO 3166-1 alpha-3': 'ISO'}, inplace = True)
cleaned_data_df.head(10)
```

```python
#We are going to drop the 'global' (ISO = 'WLD') rows because that is totals for the entire world
world_data = cleaned_data_df[cleaned_data_df['ISO'] == 'WLD']

cleaned_data_df = cleaned_data_df[cleaned_data_df['Country']!= 'Global']
```

```python
#We are going to drop the 'International Transport' (ISO = XIT) rows because that is not a real country/place
transport_data = cleaned_data_df[cleaned_data_df['ISO'] == 'XIT']

cleaned_data_df = cleaned_data_df[cleaned_data_df['Country']!= 'International Transport']
```

```python
# let's see the top 10 countries in total emissions over time
countries = list(cleaned_data_df.groupby('Country').sum().sort_values(by='Total', ascending = False).index)

print(countries)

values = list(cleaned_data_df.groupby('Country').sum().sort_values(by= 'Total', ascending=False)['Total'])

print(values)
```

```python
#Unique country
country_count = len(cleaned_data_df['Country'].unique())
country_count
```

```python
#Create dataframe with top ten countries by total emissions
top_ten_emissions_by_total = cleaned_data_df.groupby('Country').sum().sort_values(by= 'Total', ascending=False)['Total']
top_ten_emissions_by_total
```

```python
#create dataframe with top ten countires by coal emissions
top_coal = cleaned_data_df.groupby('Country').sum().sort_values(by= 'Coal', ascending=False)['Coal']
top_coal
```

```python
#create dataframe with top ten countries by oil emissions
top_oil = cleaned_data_df.groupby('Country').sum().sort_values(by= 'Oil', ascending=False)['Oil']
top_oil
```

```python
top_gas = cleaned_data_df.groupby('Country').sum().sort_values(by= 'Gas', ascending=False)['Gas']
top_gas
```

```python
top_cement = cleaned_data_df.groupby('Country').sum().sort_values(by= 'Cement', ascending=False)['Cement']
top_cement
```

```python
top_flaring = cleaned_data_df.groupby('Country').sum().sort_values(by= 'Flaring', ascending=False)['Flaring']
top_flaring
```

```python
#Create a dataframe that summarizes the top ten emissions for each category
top_ten_summary = pd.DataFrame(
    {'Total Emissions' : top_ten_emissions_by_total,
     'Coal': top_coal, 
     'Oil': top_oil,
     'Gas' : top_gas, 
     "Cement": top_cement,
     "Flaring": top_flaring})

top_ten_summary
```

```python
#Sort the new dataframe by top total emissions and display the top 10
highest_total_emissions = top_ten_summary.sort_values(['Total Emissions'], ascending= False)
highest_total_emissions.head(10)
```

```python
final_total_emissions = pd.DataFrame(highest_total_emissions)
final_total_emissions
```
